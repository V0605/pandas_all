```python
# First, create a dictionary with employee IDs and names
# where the key is the ID and the value is the name.
employees = {
12345: "Jean-Luc",
98766: "Deanna",
29384: "Geordi"
}
employees
```

```python
d = {'Time': [1, 2,3,4,5], 'EmployeeID': [98766,29384,98766,12345,29384]}
df = pd.DataFrame(data=d)
```

```python
df['Employee_Name'] = df['EmployeeID'].map(employees)
df2 = df[['Employee_Name', 'Time']]
nt = df2['Time']*5.4
df2['Time2']= nt
```

```python
df2 = df2[['Employee_Name','Time2']]
```

```python
%pwd
```

```python
alfa = %pwd
wkbks = glob(os.path.join(alfa,  'xls_files_all', 'WIC*.xls'))
sorted(wkbks)
```

```python
shts = ['Total Infants', 'Children Participating', 'Food Costs']
wic = pd.DataFrame()
```

```python
for w in wkbks:
    for s in shts:
        frame = pd.read_excel(w, sheet_name=s, skiprows=4
                  , skip_footer=4)
        frame['Type'] = s
        frame = frame.melt(id_vars=['Type', 'State Agency or Indian Tribal Organization'])
        wic = wic.append(frame, ignore_index=True)

print(wic.shape)
print(wic.columns)
```

```python
wic.head()
```

```python
#Clean up
# change columns names
wic.columns = ['Type','Area','Month','Value']
#ridding of region totals
wic = wic[(wic.Area.str.contains('Region') == False)].copy()
# cleaning dates format
wic['Month'] = pd.to_datetime(wic['Month'], errors = 'coerce')
wic= wic[wic.Month.notnull()]
# integer format
wic['Value'] = pd.to_numeric(wic['Value'], downcast='float')
wic = wic[wic.Value.notnull()]
wic.head()
```

```python
wkbks2 = glob(os.path.join(alfa , 'xls_files_all', 'est*.xls'))
sorted(wkbks2)
```

```python
pov = pd.DataFrame()
y = 2013
for w in wkbks2:
    frame = pd.read_excel(w, skiprows = 3, usecols = [2,26])
    frame['Year'] = y
    pov = pov.append(frame, ignore_index = True)
    y = y+1

```
